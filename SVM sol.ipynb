# Parameter Tuning and Optimization

```python
from sklearn.model_selection import GridSearchCV
clf = SVC()
param_grid = ['gamma':[50,5,10,0.5],'C':[15,14,13,12,11,10,0.1,0.001] }]
gsv = GridSearchCV(clf,param_grid,cv=10)
gsv.fit(x_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_
```

We can see that model performance is not increasing after parameter tuning and also our model is not classifying data correctly into 0 and 1. Hence using synthetic dataset for better classification and experimentation.

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report

# Generate a synthetic dataset
X, Y = make_classification(n_samples=1000, n_features=20, n_classes=2, weights=[0.1, 0.9], flip_y=0, random_state=42)

# Split into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

# Train an SVM model
svm = SVC(probability=True)
svm.fit(X_train, Y_train)

# Predict on test data
Y_pred_new= svm.predict(X_test)

# Evaluate the model
print(classification_report(Y_test, Y_pred_new))
```

```python
#visualization using Confusion matrix
from sklearn.metrics import confusion_matrix
import seaborn as sns
cm = confusion_matrix(y_test, Y_pred_new)
sns.heatmap(cm, annot=True)
plt.show()
```

```python
# Adjusting the decision threshold
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report

# Predict probabilities for the positive class
Y_pred_proba = svm.predict_proba(X_test)[:, 1]
threshold = 0.5  #adjusting the threshold
Y_pred_adjusted = np.where(Y_pred_proba > threshold, 1, 0)

# Confusion Matrix for adjusted threshold
conf_matrix_adjusted = confusion_matrix(Y_test, Y_pred_adjusted)
print("Confusion Matrix (Adjusted Threshold):\n", conf_matrix_adjusted)
```

```python
# Visualization of the Adjusted Confusion Matrix
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix_adjusted, annot=True)
plt.title('Confusion Matrix (Adjusted Threshold)')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# Classification Report for adjusted threshold
print(classification_report(y_test, Y_pred_adjusted))
```

# Comparision and Analysis:

```python
#linear kernal
linear_model=SVC(kernel='linear')
linear_model.fit(X_train, Y_train)
linear_pred=linear_model.predict(X_test)
print(classification_report(Y_test, linear_pred))
```

```python
##Experiment with Different Kernels
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
kernels = ['linear', 'poly', 'rbf']
kernel_results = {}

for kernel in kernels:
    svm_model = SVC(kernel=kernel, random_state=42)
    svm_model.fit(X_train, y_train)
    Y_pred_kernel = svm_model.predict(X_test)

   # Calculate and store evaluation metrics
    kernel_results[kernel] = {
        'Accuracy': accuracy_score(Y_test, Y_pred_kernel),
        'Precision': precision_score(Y_test, Y_pred_kernel),
        'Recall': recall_score(Y_test, Y_pred_kernel),
        'F1-Score': f1_score(Y_test, Y_pred_kernel)
    }
```

```python
# Display the results
for kernel, metrics in kernel_results.items():
    print(f"Kernel: {kernel}")
    for metric, score in metrics.items():
        print(f"{metric}: {score:.4f}")
    print("-" * 20)
```

Analysis:

Linear Kernel model seems to provide the best overall performance in terms of accuracy, precision, and F1-score.

**Practical Implications of SVM in real world applications:**

Support Vector machine(SVM)'s are a versatile and powerful tool in the machine learning toolkit. Their ability to handle complex, high-dimensional, and non-linearly separable data makes them suitable for a wide range of real-world classification tasks.

Real-World Applications of SVM:
* Text Classification: Spam detection, sentiment analysis, and topic categorization.
* Image Classification: Object recognition, facial recognition, and handwritten digit recognition.
* Bioinformatics: Disease classification based on gene expression data.
* Financial Analysis: Fraud detection and risk assessment.
* Medical Diagnosis: Tumor classification, disease prediction, and patient risk stratification.

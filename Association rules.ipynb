```python
import pandas as pd
from apyori import apriori

# Load the dataset
file_path = 'C:\\Users\\mukes\\Downloads\\Association Rules\\Association Rules\\Online retail.xlsx'
data = pd.read_excel(file_path)

# Rename the single column for easier access
data.columns = ['Transaction']

# Split the items in each transaction and create a list of lists
transactions = data['Transaction'].str.split(',').tolist()

# Run Apriori algorithm with specified parameters for support, confidence, and lift
# Adjust min_support, min_confidence, and min_lift based on your needs
rules = apriori(transactions, min_support=0.01, min_confidence=0.2, min_lift=3, min_length=2)

# Collect and display the results in a readable format
results = list(rules)
formatted_results = []
for result in results:
    # Each result has support, ordered statistics (confidence, lift, etc.), and items
    support = result.support
    for ordered_stat in result.ordered_statistics:
        confidence = ordered_stat.confidence
        lift = ordered_stat.lift
        items_base = tuple(ordered_stat.items_base)
        items_add = tuple(ordered_stat.items_add)
        formatted_results.append({
            'Base Items': items_base,
            'Add Items': items_add,
            'Support': support,
            'Confidence': confidence,
            'Lift': lift
        })

# Convert results to DataFrame for easier reading and display
df_results = pd.DataFrame(formatted_results)
df_results.head()
```

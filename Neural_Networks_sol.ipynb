# **Data PreProcessing**

```python
#checking for duplicates
df.duplicated().sum()
```

```python
#dropping duplicates
df.drop_duplicates(inplace=True)
```

```python
#verifying the presence of duplicates
df.duplicated().sum()
```

```python
#shape of dataset after dropping duplicates
df.shape
```

```python
#splitting data into target and features
X=df.drop('letter',axis=1)
y=df['letter']
```

```python
#normalizing data using standardscaler method
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
#scaled features
X_scaled
```

```python
#encoding target variable y using labelencoding method
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y_encoded=le.fit_transform(y)
```

```python
#target variable
y_encoded
```

# **Model Implemenation**

Constructing a basic ANN model with atleast one hidden layer using Tensorflow library Keras model.

```python
#importing tensorflow library
import tensorflow as tf
#importing sequential model and dense layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

#creating a sequential model
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(X_scaled.shape[1],)))  # First hidden layer
model.add(Dense(32, activation='relu'))  # Second hidden layer
model.add(Dense(len(set(y_encoded)), activation='softmax'))  # Output layer (for multi-class classification)

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

Splitting dataset for training and testing

```python
#training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42)
```

```python
#shape of training and testing set
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#training the model using training set
ANN=model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))
```

```python
#making predictions on test set
y_pred = model.predict(X_test)
```

```python
#predictions using basic ANN model
y_pred
```

```python
#shape of y_pred
y_pred.shape
```

```python
#getting predicted class labels for each row of y_pred
import numpy as np
y_pred_classes = np.argmax(y_pred, axis=1)
#assigning true class labels (from test set) to the variable y_true
#y_test is 1-dimensional and already has correct class labels
y_true=y_test
```

```python
# Evaluate the basic ANN model
#using classification report and accuracy score metrics
from sklearn.metrics import classification_report, accuracy_score

initial_accuracy = accuracy_score(y_true, y_pred_classes)
initial_classification_report = classification_report(y_true, y_pred_classes)

print("Initial Model Accuracy:", initial_accuracy)
print("Initial Classification Report:\n", initial_classification_report)
```

# Hyperparameter Tuning

Key Hyperparameters:

Number of hidden layers and neurons per layer

Activation functions: 'relu', 'tanh', etc.

Learning rate: Controlled by optimizer parameters

Batch size and number of epochs

Using GridSearch CV for tuning hyperparameters

```python
!pip install keras
```

```python
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV

def create_model(activation='relu', neurons=32, optimizer='adam'):
    model = Sequential()
    model.add(Dense(neurons, activation=activation, input_shape=(X_scaled.shape[1],)))
    model.add(Dense(32, activation=activation))
    model.add(Dense(len(set(y_encoded)), activation='softmax'))
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

model = KerasClassifier(model=create_model, activation='relu',neurons=32,optimizer='adam')

param_grid = {
    'activation': ['relu', 'tanh'],
    'neurons': [32, 64],
    'batch_size': [32, 64],
    'epochs': [20, 40],
    'optimizer': ['adam', 'rmsprop']
}

grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=3)
grid_result = grid.fit(X_train, y_train)
```

```python
#best accuracy value using best parameters acuqired using gridsearchsv method
print("Best accuracy: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

# **Evaluation**

Building new model with best hyperparameters and making predictions using this model.

```python
# Building the best model from Grid Search
best_model = grid_result.best_estimator_

#evaluating best model from Grid Search
y_pred_tuned = best_model.predict(X_test)
y_pred_tuned_classes = y_pred_tuned
```

Evaluating the model performance using classification report and accuracy score

```python
# Evaluate the model
#using classification report and accuracy score metrics
from sklearn.metrics import classification_report, accuracy_score

initial_accuracy = accuracy_score(y_true, y_pred_tuned_classes)
initial_classification_report = classification_report(y_true, y_pred_tuned_classes)

print("Tuned Model Accuracy:", initial_accuracy)
print("Tuned Model Classification Report:\n", initial_classification_report)
```

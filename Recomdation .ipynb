Feature Extraction :

```python
# Convert categorical features into numerical representations if necessary.
# Normalize numerical features if required.

from sklearn.preprocessing import MultiLabelBinarizer

# Feature Selection: Genres and Rating
features_df = df[['genre', 'rating']]

# Convert 'genre' to numerical representation using one-hot encoding
mlb = MultiLabelBinarizer()
genre_encoded = pd.DataFrame(mlb.fit_transform(features_df['genre'].str.split(',')),
                             columns=mlb.classes_,
                             index=features_df.index)

features_df = pd.concat([features_df, genre_encoded], axis=1)
features_df.drop('genre', axis=1, inplace=True)


# Normalize 'rating' (if necessary) - you can choose to normalize or not
# You might want to consider the range of ratings in your dataset and whether it needs normalization
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
features_df['rating'] = scaler.fit_transform(features_df[['rating']])

print(features_df.head())
```

Recommendation System:

```python
# Given a target anime, recommend a list of similar anime based on cosine similarity scores.
# Experiment with different threshold values for similarity scores to adjust the recommendation list size.

from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def recommend_anime(anime_title, features_df, similarity_threshold=0.7):
  """
  Recommends anime based on cosine similarity.

  Args:
    anime_title: The title of the target anime.
    features_df: A DataFrame with anime features (e.g., genre, rating).
    similarity_threshold: The minimum cosine similarity score for a recommendation.

  Returns:
    A list of recommended anime titles.
  """

  if anime_title not in df['name'].values:
    return "Anime not found in the dataset."

  target_anime_index = df[df['name'] == anime_title].index[0]
  target_anime_features = features_df.iloc[target_anime_index].values.reshape(1, -1)

  # Check for NaN values in target_anime_features and features_df and handle them (e.g., impute with 0)
  target_anime_features = np.nan_to_num(target_anime_features)
  features_df = features_df.fillna(0)

  cosine_similarities = cosine_similarity(target_anime_features, features_df)
  similar_anime_indices = np.where(cosine_similarities > similarity_threshold)[1]
  similar_anime_scores = cosine_similarities[0][similar_anime_indices]

  recommended_anime = []
  for index, score in zip(similar_anime_indices, similar_anime_scores):
    recommended_anime.append((df['name'].iloc[index], score))

  # Sort recommendations by similarity score in descending order
  recommended_anime.sort(key=lambda x: x[1], reverse=True)

  return [anime[0] for anime in recommended_anime]

# Example usage:
target_anime = "Naruto"
recommendations = recommend_anime(target_anime, features_df)
print(f"Recommendations for '{target_anime}':")
for anime in recommendations:
  print(anime)
```

```python
def evaluate_recommendations(df, features_df, recommend_anime_func, test_size=0.2, random_state=42):
  """
  Evaluates the recommendation system using precision, recall, and F1-score.

  Args:
    df: The original DataFrame with anime data.
    features_df: The DataFrame with features used for recommendations.
    recommend_anime_func: The function that generates recommendations.
    test_size: The proportion of the dataset to include in the test split.
    random_state: The seed used by the random number generator.

  Returns:
    A dictionary containing precision, recall, and F1-score.
  """

  train_df, test_df = train_test_split(df, test_size=test_size, random_state=random_state)

  true_positives = 0
  false_positives = 0
  false_negatives = 0

  for index, row in test_df.iterrows():
    anime_title = row['name']
    recommendations = recommend_anime_func(anime_title, features_df)

    if isinstance(recommendations, str):  # Handle cases where anime is not found
      continue

    # Check if any recommended anime is also in the test set
    for recommended_anime in recommendations:
      if recommended_anime in test_df['name'].values:
        true_positives += 1
        break
    else:
      false_negatives += 1

    # Check for false positives (recommendations not in the test set) - you can also consider a threshold for this
    # ...

  precision = true_positives / (true_positives + false_positives) if (true_positives + false_positives) > 0 else 0
  recall = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) > 0 else 0
  f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

  return {'precision': precision, 'recall': recall, 'f1_score': f1_score}


# Evaluate the recommendation system
evaluation_results = evaluate_recommendations(df, features_df, recommend_anime)
print("Evaluation Results:")
print(evaluation_results)
```

Interview Questions:

1. Can you explain the difference between user-based and item-based collaborative filtering?

User-Based Collaborative Filtering:

How it Works: Recommends items to you based on what similar people like.*

Example: If you and a friend liked the same movies, and your friend found a new movie they loved, that movie might be suggested to you.*

Item-Based Collaborative Filtering:

How it Works: Recommends items similar to ones you’ve liked before.*

Example: If you liked a specific movie, it might suggest other movies that people who liked that movie also enjoyed.*

Main Differences:

Focus: User-based looks at people; item-based looks at items.*

Scalability: Item-based is usually easier to scale since there are often fewer items than users.*

Cold Start: User-based struggles with new users; item-based can manage new items better.*


2. What is collaborative filtering, and how does it work?

**What is Collaborative Filtering?**

*Collaborative filtering is a way to recommend things (like movies or music) to you based on what other people with similar tastes like.*

**How It Works:**

**Collect Data:** Gather information on what users like—this can be ratings, clicks, or purchases.*

**Find Similarities:**

**User-Based:** Look for people who have similar likes to you.*

**Item-Based:** Look for items that are liked by the same group of people.*

**Make Recommendations:**

**User-Based:** If someone similar to you liked a movie you haven’t seen, it gets suggested to you.*

**Item-Based:** If you liked a movie, it will recommend other movies that people who liked your movie also enjoyed.*
